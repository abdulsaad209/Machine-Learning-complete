```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
%matplotlib inline
```

```python
X,y = make_classification(n_samples=5, n_features=2, n_classes=2, n_informative=2, n_redundant=0, random_state=0)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
clf.feature_importances_ # sum of all features importance should be 1
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = DecisionTreeClassifier()
```

```python
rf = RandomForestClassifier(n_estimators=2) # i just take 2 decision tree as base model
```

```python
X,y = make_classification(n_samples=5, n_features=2, n_classes=2, n_informative=2, n_redundant=0, random_state=0)
```

```python
rf.fit(X,y)
```

```python
rf.feature_importances_
```

```python
print(rf.estimators_[0].feature_importances_)
print(rf.estimators_[1].feature_importances_)

print((1+0.625)/2)
```
