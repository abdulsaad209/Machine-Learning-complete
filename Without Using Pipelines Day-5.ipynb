# *Titanic Without using pipeline*
Model Training

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/home/saad/Downloads/Titanic.csv')
df = df.iloc[:,1:]
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'], inplace=True)
```

```python
df.head()
```

```python
# 1 step -> train test split
X_train,X_test,y_train,y_test=train_test_split(df.drop('Survived',axis=1),df['Survived'],
                                               test_size=0.2,random_state=True)
```

```python
X_train.shape
```

```python
df.isnull().sum()
```

```python
# Applying Imputation

si_age = SimpleImputer()
#si_fare = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])
#X_train_Fare = si_fare.fit_transform(X_train[['Fare']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
#X_test_Fare = si_fare.transform(X_test[['Fare']])
```

```python
X_train_embarked
```

X_test_embarked = si_embarked.transform(X_test[['Embarked']])

```python
# One hot Encoding on Sex & Embarked

ohe_sex = OneHotEncoder(sparse=False, handle_unknown='ignore', dtype='int64')
ohe_embarked = OneHotEncoder(sparse=False, handle_unknown='ignore', dtype='int64')
# i create two objects for Encoding
# because there are still null values in the original data set in Embarked and Age features


X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked)
```

```python
X_train_sex
```

```python
X_train_embarked
```

```python
# remaining feature
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_sex,X_train_embarked),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_sex,X_test_embarked),axis=1)
```

```python
X_train_transformed.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
